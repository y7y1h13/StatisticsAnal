```python
import pandas as pd

df1 = pd.read_csv('data/2007~2011노인교통사고(서울).csv')
df2 = pd.read_csv('data/2012~2016노인교통사고(서울).csv')
df3 = pd.read_csv('data/2017~2020노인교통사고(서울).csv')
```

```python
df2.drop('시군구',axis=1,inplace=True)
df3.drop('시군구',axis=1,inplace=True)
```

```python
df = pd.concat([df1,df2,df3],axis=1)
```

```python
df
```

```python
df = df.T
df
```

```python
df = df.rename(columns=df.iloc[0])
df
```

```python
df = df.drop(df.index[0])
df
```

```python
df.reset_index(inplace=True)
df
```

```python
df.rename(columns={'index':'지정일'},inplace=True)
```

```python
df.info()
```

```python
df = df.astype({'지정일':'int64'})
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data2/서울특별시 경로당 정보.csv')
```

```python
df.head()
```

```python
df.drop(['번호','도로명전체주소','인허가일자','영업상태명','폐업일자','휴업시작일자','휴업종료일자','재개업일자','소재지면적','입소정원','소재지우편번호','자격소유인원수','총인원수','인허가번호','상세영업상태명'],axis=1,inplace=True)
```

```python
df
```

```python
n = []
for i in range(len(df)):
    if '동대문구' in df.loc[i,'소재지전체주소']:
        n.append(df.loc[i])
n
```

```python
df_1 = pd.DataFrame(n)
```

```python
df_1
```

```python
df_1.reset_index(inplace=True)
```

```python
df_1.drop('index',axis=1,inplace=True)
```

```python
df_1
```

```python
df_1.columns = ['이름','주소','x','y']
```

```python
df_1['주소'] = df_1['주소'].str.replace('서울특별시 ','')
```

```python
df_1.head()
```

```python
df_1.to_csv('data2/동대문구 경로당.csv')
```

```python
def uniqu(s):
    a = []
    for i in s:
        if i not in a:
            a.append(i)
    a.sort()
    return a
#데이터 전달할때는 정렬까지 해서 줘라
# def u_unique(data):
#     udata=[]
    
#     for i in data:
#         if i not in udata:
#             udata.append(i)
#     return udata
```

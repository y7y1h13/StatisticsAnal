```python
plt.rcParams["figure.figsize"] = (14,7)
plt.rcParams['font.size'] = 12
ax = all.plot.bar(y=['노인','어린이'],rot=0,color=['r','c'],width=0.8)
for temp in ax.patches:
    ax.annotate('%d'%temp.get_height(),(temp.get_x()+temp.get_width()/2 , temp.get_height()) , ha = 'center' , va='bottom',fontsize=13)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (14,7)
plt.rcParams['font.size'] = 12
bar_width =0.4
index = np.arange(len(all.index))
ax1 = plt.bar(index,all['노인'],color='r',align='edge',width=bar_width,label='노인')
ax2 = plt.bar(index + bar_width,all['어린이'],color='c',align='edge',width=bar_width,label='어린이')


plt.xticks(index+bar_width,all.index)
plt.legend()
plt.xlabel('유형')
plt.title('노인, 어린이 보호구역 단속카메라 현황')
```

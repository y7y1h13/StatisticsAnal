#  강사님답

```python
x = np.linspace(-1,1,100)
```

```python
y = x*x

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x,y,color='red')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x=np.arange(1,11)
y=2*x+3
```

```python
plt.figure(figsize=(8,3))
plt.scatter(x,y,marker='s')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

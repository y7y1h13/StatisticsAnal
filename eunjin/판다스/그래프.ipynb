```python
import seaborn as sns

anscombe = sns.load_dataset('anscombe')
print(anscombe)
print(type(anscombe))
```

```python
# 그리기 임포트
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I'] #dataset이 I인 것만 추출
```

```python
plt.plot(dataset_1['x'], dataset_1['y']) # (x, y)
```

```python
plt.plot(dataset_1['x'], dataset_1['y'],'o') # o는 점을 나타낸다.
```

### 1. 전체 그래프가 위치할 기본 틀을 만든다.
### 2. 그래프를 그려 넣을 그래프 격자를 만든다.
### 3. 격자에 그래프를 하나씩 추가한다. 그래프 추가되는 순서는 왼쪽부터
### 4. 격자의 첫 번째 행이 꽉 차면 두 번째 행에 그래프를 그려 넣는다.

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']#데이터 저장
```

```python
fig = plt.figure()#기본틀
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)
# 가로, 세로, 위치


axes1.plot(dataset_1['x'], dataset_1['y'],'o')
axes2.plot(dataset_2['x'], dataset_2['y'],'o')
axes3.plot(dataset_3['x'], dataset_3['y'],'o')
axes4.plot(dataset_4['x'], dataset_4['y'],'o')
```

```python
axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')
#제목
fig
```

```python
fig.suptitle('Anscombe Data') # 큰 제목
fig
```

```python
fig.tight_layout()# 레이아웃 조절
fig
```

### 다양한 그래프

```python
import seaborn as sns
tips = sns.load_dataset('tips')
# 그리기 임포트
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
print(tips)
print(type(tips))
#지불 금액, 팁, 성별, 흡연 유무, 요일, 시간, 전체 인원 정보
```

```python
#히스토그램은 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도로 사용
```

```python
# 틀과 격자 생성
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'],bins=10) #bins=10,x축 간격을 10으로 조정(도수분포표) / 0~10, 10~15, 15~20, 25~30....
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
fig
```

```python
#산점도 그래프
#산점도 그래프는 변수 2개를 이용함
#두 변수의 상관 관계
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

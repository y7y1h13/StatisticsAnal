```python
import pandas as pd
```

```python
#코로나 데이터 불러옴
corona = pd.read_csv('../data2/covid_seoul.csv')
corona.head()
```

```python
#확진일만 필요함
corona_df = corona[['확진일']]
```

```python
corona_df.info()
```

```python
corona_df['확진일'] = corona_df['확진일'].str.replace('-','')
```

```python
#확진일 중에서도 년-월 데이터만 필요하기 때문에 월 데이터를 없앤다.
for i in range(len(corona_df['확진일'])):
    tmp = corona_df.iloc[i].values[0]
    corona_df.iloc[i] = tmp[:-2]
```

```python
# 확진일 데이터를 비교하여 같은 달이면 +1을 해주고 딕셔너리에 넣는다.
num = {}
for i in corona_df['확진일']:
    if i not in num:
        num[i] = 1
    else:
        num[i] += 1
num
```

```python
#데이터 프레임으로 만들어준다.
corona_df2 = pd.DataFrame(num, index=[0])
#전치행렬을 적용해준다.
corona_df2 = corona_df2.transpose()
#인덱스를 새로 만들어준다.
corona_df2.reset_index(inplace=True)
#컬럼의 이름을 바꿔준다.
corona_df2.columns = ['날짜','확진자수']
# 내림차순 정렬을 해준다.
corona_df2.sort_values(by=['날짜'], inplace=True, ascending=True)
# 인덱스를 새로만든다.
corona_df2.reset_index(inplace=True)
# 구 인덱스를 삭제한다.
corona_df2.drop('index',axis=1,inplace=True)
```

# 1. 서울시 범죄데이터를 불러오세요

```python
import pandas as pd

crimedf = pd.read_csv('./data/crimeSeoul.csv' , sep=',' , engine='python' , encoding='euc-kr' )
crimedf
```

```python
policedf = pd.read_csv('./data/police.csv' , sep=',' , engine='python', encoding='euc-kr')
policedf.head()
```

```python
policedf.info()
```

```python
policedf1 = policedf.loc[policedf['청']=='서울청',:]
policedf1
```

```python
nameresult = []
for temp in crimedf['관서명']:
    nameresult.append('서울'+temp[:-1]+"경찰서")
    
crimedf['관서명'] = nameresult
crimedf
```

```python
gudata = []

for temp in crimedf['관서명']:
    gudata.append( policedf1.loc[policedf1['지구대파출소']==temp , :].주소.str.split(' ').values[0][0] )
    
gudata
```

```python
gudata = []
for temp in crimedf['관서명']:
    tempa = policedf1.loc[policedf1['지구대파출소']==temp , :].주소
    tempb =  tempa.str.split(' ').values
    gudata.append(tempb[0][0])
gudata    
```

범죄데이터에 구컬럼을 추가하여 구데이터를 입력하세요

```python
crimedf['구'] = gudata
```

```python
crimedf.head()
```

```python
crimedf.info()
```

```python
crimedf['절도 발생'] = crimedf['절도 발생'].str.replace(',','')
crimedf['절도 발생'] = crimedf['절도 발생'].astype(int)
crimedf['절도 검거'] = crimedf['절도 검거'].str.replace(',','')
crimedf['절도 검거'] = crimedf['절도 검거'].astype(int)
crimedf['폭력 발생'] = crimedf['폭력 발생'].str.replace(',','')
crimedf['폭력 발생'] = crimedf['폭력 발생'].astype(int)
crimedf['폭력 검거'] = crimedf['폭력 검거'].str.replace(',','')
crimedf['폭력 검거'] = crimedf['폭력 검거'].astype(int)
```

```python
crimedf.info()
```

```python
crimedf = crimedf.groupby('구').sum()
crimedf
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/보행고령자 사고다발/보행고령자 사고다발지역.csv',)
```

```python
df.head()
```

```python
df_1 = df[['시도시군구명','지점명','발생건수','사상자수','경도','위도']]
```

```python
df_1
```

```python
acc = []
for i in range(len(df_1)):
    if '서울' in df_1.loc[i,'시도시군구명']:
#         print(df_1.loc[i,'시도시군구명'])
        acc.append(df_1.loc[i])
acc
```

```python
df_acc = pd.DataFrame(acc)
```

```python
df_acc
```

```python
df_acc.columns = ['구','지점','발생건수','사상자수','경도','위도']
```

```python
df_acc.head()
```

```python
df_acc['구'] = df_acc['구'].str.replace('서울특별시 ','')
```

```python
df_acc.head()
```

```python
import re
gu = []
for i in df_acc['구']:
    tmp = re.sub(r'[0-9]+','',i)
    gu.append(tmp)
gu
```

```python
df_acc['구'] = gu
```

```python
df_acc
```

```python
df_acc['지점'] = df_acc['지점'].str.replace('서울특별시 ','')
```

```python
df_acc.head()
```

```python
df_acc.reset_index(inplace=True)
```

```python
df_acc.head()
```

```python
df_acc.drop('index',axis=1,inplace=True)
```

```python
df_acc.to_csv('../data/보행고령자 사고다발/서울 고령자사고다발지역.txt')
```

```python
df_acc.head()
```

```python
j = df_acc[df_acc['구']=='동대문구']
```

```python
j.to_csv('../data/보행고령자 사고다발/동대문구 사고다발지역.csv')
```

```python
k = k.T
```

```python
k.head(10)
```

```python
k.to_csv('../data/보행고령자 사고다발/구별 사고 발생건수.csv')
```

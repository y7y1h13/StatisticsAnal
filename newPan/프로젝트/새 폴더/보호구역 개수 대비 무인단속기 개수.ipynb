```python
import warnings
warnings.filterwarnings(action='ignore') 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib
import platform
import seaborn as sns 
import matplotlib
matplotlib.rcParams['font.family']
if platform.system() == 'Windows':
    matplotlib.rc('font', family='Malgun Gothic')
    #Malgun 
%matplotlib inline
```

```python
df1 = pd.read_csv('./년도별 어린이 보호구역 합계 (1).csv')
df2 = pd.read_csv('./어린이무인단속장비 현황 (1).csv')
```

```python
df1
```

```python
a = df1[['2021']]
```

```python
a
```

```python
aa = a.iloc[1:2]
aa
```

```python
df2
```

```python
b = df2[['계']]
b
```

```python
bb = b.iloc[0:1]
bb
```

```python
aa.columns = ["어린이 보호구역"]
bb.columns = ["어린이 보호구역 단속기"]
print (aa)
print (bb)
```

```python
aa.index = ['0']
aa
```

```python
bb.index = ['0']
bb
```

```python
ab = pd.concat([aa,bb], axis=1)
ab
```

```python
ab = ab.rename({'0':'갯수'}, axis=0)
ab
```

```python
ab = ab.T
ab
```

```python
ab.index
```

```python
ax = ab.plot(kind='bar',rot=360,fontsize=8,width=0.4, color=['slateblue', 'darkslateblue'])
plt.title("어린이 보호구역과 단속기 비교")
plt.xlabel("비교")
for temp in ax.patches:
    ax.annotate('%d'%temp.get_height(),(temp.get_x()+temp.get_width()/2 , temp.get_height()) , ha = 'center' , va='bottom',fontsize=8)
plt.show()
```

```python
ax = ab.plot(kind='barh',rot=360,fontsize=8,width=0.4, color=['slateblue', 'darkslateblue'])
plt.title("어린이 보호구역과 단속기 비교")
plt.xlabel("비교")
plt.show()
```

```python
plt.pie(ab['갯수'], labels=ab.index, autopct='%1f%%', startangle=200)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../data/gdp.csv', index_col=0)
df
```

```python
all = df.loc[:,['Korea_Rep','Japan','United_Kingdom','Germany','United_States']]
```

```python

ax = plt.subplots()
ax = plt.plot(all)
```

```python
axe=df[['Korea_Rep','United_States','United_Kingdom','Germany','China','Japan']].plot()
axe.set_title("Each country's GDP")
axe.set_xlabel('YEAR',loc='right')
axe.set_ylabel('GDP',loc='top',rotation='horizontal')
axe.set_xlim(2000,2017)
```

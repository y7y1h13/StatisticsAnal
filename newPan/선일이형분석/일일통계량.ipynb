```python
traffic2001['총교통량'] = traffic2001['0시']+traffic2001['1시']+traffic2001['2시']+traffic2001['3시']+traffic2001['4시']+traffic2001['5시']+traffic2001['6시']+traffic2001['7시']+traffic2001['8시']+traffic2001['9시']+traffic2001['10시']+traffic2001['11시']+traffic2001['12시']+traffic2001['13시']+traffic2001['14시']+traffic2001['15시']+traffic2001['16시']+traffic2001['17시']+traffic2001['18시']+traffic2001['19시']+traffic2001['20시']+traffic2001['21시']+traffic2001['22시']+traffic2001['23시']
traffic2001['일자'] = pd.to_datetime(traffic2001['일자'], format='%Y%m%d')
traffic2001.head(2)
```

```python
t20_01 = traffic2001[['일자', '요일', '지점명', '방향', '총교통량']]
t20_01
```

```python
t2001_inout = t20_01.groupby(['일자', '요일'])['총교통량'].sum()
t2001_inout
```

```python
df_2001 = pd.DataFrame(t2001_inout)
df_2001
```

```python
df_2001.to_csv('data/traffic/202001_202107.txt')
```

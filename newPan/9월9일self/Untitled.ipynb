```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent' : 'mozilla/5.0', 'Accept-Encoding':'gzip,deflate','Accept':'*/*','Connection':'keep-alive'}
page = requests.get('https://news.naver.com', headers = headers)
html = page.text
html
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
data1 = soup.find('div', {'class' : 'hdline_news'})
data1
# result = bs4.find('div', id='ex_id')
```

```python
data2 = data1.find('ul',{'class':'hdline_article_list'})
data2
```

```python
data3 = data2.find_all('div',{'class' : 'hdline_article_tit'})
data3
```

```python
soup = BeautifulSoup(html,'lxml')
temp = soup.find('ul',{'class':'hdline_article_list'})
temp1 = temp.find_all('div',{'class':'hdline_article_tit'})

titlelist = []
for temp2 in temp1:
    titlelist.append(temp2.text.strip())

titlelist
```

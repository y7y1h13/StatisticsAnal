남/여 값을 이용하여 남=1, 여=2 로 변환해서 성별코드 컬럼을 새로 생성해서 남/여와 학년사이에 위치하도록 수정하세요.

```python
df['성별코드'] = [ 1 if i == '남자' else 2  for i in df['남/여']]
df.head()
```

```python
df1['학과'].unique()
```

학과의 고유값을 출력해주는 함수를 작성해보세요.

```python
def u_unique(data):
    udata = []
    
    for i in data:
        if i not in udata:
            udata.append(i)

    udata.sort()
    return udata

u_unique(df1['학과'])

    
```

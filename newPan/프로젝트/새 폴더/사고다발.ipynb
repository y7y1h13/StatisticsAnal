```python
import pandas as pd
```

```python
df = pd.read_csv('./보행고령자 사고다발지역.csv',encoding='cp949')
```

```python
df.head()
```

```python
df_1 = df[['시도시군구명','지점명','발생건수','사상자수','경도','위도']]
```

```python
df_1
```

```python
acc = []
for i in range(len(df_1)):
    if '서울' in df_1.loc[i,'시도시군구명']:
#         print(df_1.loc[i,'시도시군구명'])
        acc.append(df_1.loc[i])
acc
```

```python
df_acc = pd.DataFrame(acc)
```

```python
df_acc
```

```python
df_acc.columns = ['구','지점','발생건수','사상자수','경도','위도']
```

```python
df_acc.head()
```

```python
df_acc['구'] = df_acc['구'].str.replace('서울특별시 ','')
```

```python
df_acc.head()
```

```python
import re
gu = []
for i in df_acc['구']:
    tmp = re.sub(r'[0-9]+','',i)
    gu.append(tmp)
gu
```

```python
df_acc['구'] = gu
```

```python
df_acc
```

```python
df_acc['지점'] = df_acc['지점'].str.replace('서울특별시 ','')
```

```python
df_acc.head()
```

```python
df_acc.reset_index(inplace=True)
```

```python
df_acc.head()
```

```python
df_acc.drop('index',axis=1,inplace=True)
```

```python
# df_acc.to_csv('data2/서울 고령자사고다발지역.csv')
```

```python
df_acc.head()
```

```python
import folium

center = [37.541 , 126.986]
mapdata = folium.Map(location=center, zoom_start = 12)

for temp in df_acc.index:
    folium.Circle(
    [df_acc['위도'][temp], df_acc['경도'][temp]],
    radius = int(df_acc.loc[temp, ['발생건수']].values[0] * 10), color='#3185cc', fill_color='#3185cc'
    
    ).add_to(mapdata)
mapdata
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform
if platform.system() == 'Windows':
    matplotlib.rc('font', family='Malgun Gothic')

%matplotlib inline
```

```python
covid = pd.read_csv('data/covid_seoul1.txt', encoding='ANSI')
covid.head()
```

```python
covid1 = covid.drop(['환자번호', '국적', '환자정보', '지역', '여행력', '접촉력', '조치사항', '상태', '이동경로', '등록일', '수정일', '노출여부'], axis=1)
```

```python
covid1
```

```python
covid1['확진일'] = pd.to_datetime(covid1['확진일'], format='%Y-%m-%d')
covid1
```

```python
covid1.info()
```

```python
covid1_month_count = covid1.groupby(['확진일']).size()
covid1_month_count
```

```python
covid_day_count = pd.DataFrame(covid1_month_count, columns=['일별 확진자 수'])
```

```python
covid_day_count
```

```python
covid_day_count.to_csv('data/covid_day_count.txt')
```

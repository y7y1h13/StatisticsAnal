```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import matplotlib
import platform
if platform.system() == 'Windows':
    matplotlib.rc('font', family='Malgun Gothic')
    #Malgun 
%matplotlib inline
```

```python
df = pd.read_csv('data2/서울시 구별 고령자 인구수.csv')
```

```python
df.head()
```

```python
df = df[['지역','계']]
```

```python
df.drop(0,inplace=True)
```

```python
df['계'] = df['계'].str.replace(',','')
```

```python
df['계'] = df['계'].astype('int')
```

```python
df
```

```python
plt.rcParams["figure.figsize"] = (14,7)
plt.rcParams['font.size'] = 12
plt.pie(df['계'],labels=df['지역'])
plt.show()
```

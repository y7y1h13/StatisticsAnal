```python
dataset = [
    ['커피','녹차','얼음'],   #1인구매 리스트
    ['커피','얼음','우유'],
    ['커피','딸기'],
    ['바나나','녹차','딸기'],
    ['커피','얼음'],
    ['우유','커피'],
    ['우유','딸기'],
    ['커피'],
    ['얼음','커피']
]

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
f_data = apriori(df, min_support = 0.2 , use_colnames=True)
f_data
```

```python
from mlxtend.frequent_patterns import association_rules
#지지도(support) 전체거래항목중 a,b를 포함하는 비율
#신뢰도(confidence) a를 포함하는 거래중에 a,b가 동시에 거래되는 비중
#향상도(Lift) a와 b사이에 상호관계가 없으면 1     1보다 큰값 연관성이 높다   1보다 작은값 연관성이 낮다

association_rules(f_data , metric='lift' , min_threshold = 1)
```

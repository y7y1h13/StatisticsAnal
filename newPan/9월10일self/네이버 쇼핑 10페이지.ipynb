```python
import requests
from bs4 import BeautifulSoup
import json
import urllib 
import time

keyword = input('네이버 쇼핑 검색어 ')
keyword = urllib.parse.quote(keyword)
result = []
for i in range(1,11):
    page = requests.get('https://search.shopping.naver.com/search/all?frm=NVSHATC&pagingIndex={0}&pagingSize=40&productSet=total&query={1}'.format(i,keyword))
    html = page.text

    soup = BeautifulSoup(html,'lxml')

    jsonresult = soup.find('script',{'id':'__NEXT_DATA__'})
    jsondata = json.loads(jsonresult.string)


    jdata1 = jsondata['props']['pageProps']['initialState']['products']['list']


    for temp in jdata1:
        rtemp = []
        rtemp.append(temp['item']['productTitle'])
        rtemp.append(temp['item']['lowPrice'])

        result.append(rtemp)

f = open('{0}-네이버 쇼핑검색.csv'.format(keyword),'w')
f.write('상품이름'+','+'최저가'+'\n')
for temp in result:
    f.write(str(temp[0]) + ',' + str(temp[1]) + '\n')
    
f.close()
```

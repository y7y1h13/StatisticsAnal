```python
result = []
for temp in bdata:
    result.append(temp[2])
    
```

```python
spliter = Okt()
result_list = []

for temp in result:
    nouns = spliter.nouns(temp)
    result_list.append(nouns)
    
result_list
```

```python
import gensim
import gensim.corpora as corpora

dictionary = corpora.Dictionary(result_list)

corpus = []

for temp in result_list:
    corpus.append(dictionary.doc2bow(temp))
    
corpus
```

```python
dictionary.token2id
```

```python
k = 10  #토픽의 갯수

lda_model = gensim.models.ldamulticore.LdaMulticore(corpus,iterations = 24, num_topics=k, id2word=dictionary, passes = 2, workers=10 )

print(lda_model.print_topics(num_topics=k, num_words=15))
```

```python
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

pyLDAvis.enable_notebook()
lda_vis = gensimvis.prepare(lda_model , corpus , dictionary)
pyLDAvis.display(lda_vis)
```

```python
import time
from gensim.models.coherencemodel import CoherenceModel

coherences=[]
perplexities=[]
passes=[]


for i in range(10):
    
    ntopics, nwords = 200, 100
    if i==0:
        p=1
    else:
        p=i*5
    tic = time.time()
    lda4 = gensim.models.ldamulticore.LdaMulticore(corpus, id2word=dictionary, num_topics=ntopics, iterations=400, passes=p)
    print('epoch',p,time.time() - tic)
    # tfidf, corpus 무슨 차이?
    # lda = models.ldamodel.LdaModel(corpus, id2word=dictionary, num_topics=ntopics, iterations=200000)

    cm = CoherenceModel(model=lda4, corpus=corpus, coherence='u_mass')
    coherence = cm.get_coherence()
    print("Coherence",coherence)
    coherences.append(coherence)
    print('Perplexity: ', lda4.log_perplexity(corpus),'\n\n')
    perplexities.append(lda4.log_perplexity(corpus))
    
```

```python
import requests
from bs4 import BeautifulSoup
import json
import urllib
import time
```

```python
keyword = input('키워드 : ')
keyword1 = urllib.parse.quote(keyword)

page = requests.get('https://search.shopping.naver.com/search/all?query={0}&frm=NVSHATC'.format(keyword1))
html = page.text
html
```

```python
soup = BeautifulSoup(html,'lxml')

jsonresult = soup.find('script',{'id' : '__NEXT_DATA__'})
```

```python
jsonresult
```

```python
jsondata = json.loads(jsonresult.string)
jdata1 = jsondata['props']['pageProps']['initialState']['products']['list']
```

```python
result = []

for temp in jdata1:
    rtemp = []
    rtemp.append(temp['item']['productTitle'])
    rtemp.append(temp['item']['lowPrice'])
    result.append(rtemp)
result
```
